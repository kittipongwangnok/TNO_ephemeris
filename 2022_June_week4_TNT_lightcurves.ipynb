```python
#################################################################################
'''
1. Input file: lcurve_dpleo_data
'''
#################################################################################
#Read data ---#4
InputFile_4 = "dpleo_20200121_run036g.dat"
Data_4   = np.genfromtxt(InputFile_4)

#Read data ---#2
InputFile_2 = "dpleo_20200121_run057r.dat"
Data_2   = np.genfromtxt(InputFile_2)

#Read data ---#16
InputFile_16 = "dpleo_20200322_run026kg5.dat"
Data_16   = np.genfromtxt(InputFile_16)
```

```python
#DP Leo parameters
#We are using the ephemeris of TO and Period obtained by Beuermann et al. (2011)
t0 = 2454914.8322920
Period = 0.06236285648

#Orbital phase calculation ---#4
Cycle_4 = (Data_4[:,0] - t0)/Period
E_4 = int(Cycle_4[0])
Phase_4 = Cycle_4 - E_4
Flux_4 = Data_4[:,2]
Flux_err_4 = Data_4[:,3]

#Orbital phase calculation ---#2
Cycle_2 = (Data_2[:,0] - t0)/Period
E_2 = int(Cycle_2[0])
Phase_2 = Cycle_2 - E_2
Flux_2 = Data_2[:,2]
Flux_err_2 = Data_2[:,3]

#Orbital phase calculation ---#16
Cycle_16 = (Data_16[:,0] - t0)/Period
E_16 = int(Cycle_16[0])
Phase_16 = Cycle_16 - E_16
Flux_16 = Data_16[:,2]
Flux_err_16 = Data_16[:,3]
```

```python
offset_4 = 0
offset_2 = 0.05
offset_16 = 0.1

fig=plt.figure(figsize=(10, 5), tight_layout=True)
plt.xlim(0.8, 1.2)
plt.ylim(-0.01, 0.16)
plt.xlabel('Orbital phase')
plt.ylabel('Relative flux + offset')

plt.tick_params(direction='in', which='both', bottom='on',top='on', right = 'on')

plt.errorbar(Phase_4, Flux_4+offset_4, yerr=Flux_err_4, fmt='o', markersize = 3, alpha = 1.00, color='blue',
                    ecolor='lightgray', elinewidth=2, capsize=0)
plt.errorbar(Phase_2, Flux_2+offset_2, yerr=Flux_err_2, fmt='o', markersize = 3, alpha = 1.00, color='red',
                    ecolor='lightgray', elinewidth=2, capsize=0)
plt.errorbar(Phase_16, Flux_16+offset_16, yerr=Flux_err_16, fmt='o', markersize = 3, alpha = 1.00, color='orange',
                    ecolor='lightgray', elinewidth=2, capsize=0)


plt.text(1.123, 0.04, '20200121\_$g^{\prime}$')
plt.text(1.123, 0.08, '20200121\_$r^{\prime}$')
plt.text(1.123, 0.14, '20200322\_KG5')

plt.grid(linestyle='dotted')

#plt.legend(loc="best")
#plt.set_ylabel('Relative flux')

fig.align_ylabels()
#output_filename = os.path.splitext(__file__)[0] + '.png'
#plt.savefig(output_filename, dpi=1000)
plt.savefig("2022_June_week4_TNT_lightcurves.pdf", dpi=1000)
plt.show()
```
